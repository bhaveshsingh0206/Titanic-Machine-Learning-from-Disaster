```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
X_train[:,2:3] = imp.fit_transform(X_train[:,2:3])
X_test[:,2:3] = imp.fit_transform(X_test[:,2:3])
X_test[:,5:6] = imp.fit_transform(X_test[:,5:6])
```

```python
# from sklearn.preprocessing import LabelEncoder
# labelencoder = LabelEncoder()
# X_train[:,7] = labelencoder.fit_transform(X_train[:,7])
```

```python
X_train = X_train[:,1:]
X_test = X_test[:,1:]
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
####################LogisticRegression####################
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
y_pred_log = classifier.predict(X_test)
```

```python
####################KNN Neighbour####################
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
y_pred_knn = classifier.predict(X_test)
```

```python
####################GaussianNB####################
from sklearn.naive_bayes import GaussianNB
classifier_nb = GaussianNB()
classifier_nb.fit(X_train, y_train)
y_pred_bayes = classifier.predict(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_tr = sc_X.fit_transform(X_train)
X_te = sc_X.transform(X_test)
```

```python
####################Kernel SVM####################
from sklearn.svm import SVC
classifier_svm = SVC(kernel='rbf')
classifier_svm.fit(X_tr, y_train)
y_pred_svm = classifier.predict(X_te)
```

```python
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score(classifier_nb, X_train, y_train, cv=10)
accuracy.mean()
```
